```python
# for CSV
import pandas as pd
import numpy as np

# for Data Preprocessing
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt_tab')
from bs4 import BeautifulSoup
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler

# for Neural Network Model
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB  # ('bag of words' model), for comparison?
from sklearn.feature_extraction.text import TfidfVectorizer

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers # Keras used for model
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# For evaluating learning model
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

# For graph plotting
import matplotlib.pyplot as plt
from sklearn.model_selection import LearningCurveDisplay, ShuffleSplit
```

```python
# Drop the csv file into your Google Drive (or local)
# Read the csv file

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/IMDB Dataset.csv')
# For local use
#data = pd.read_csv('/IMDB Dataset.csv')

# Summary of dataset
#data.head(5)
data.describe()
```

```python
data['sentiment'].value_counts()
```

```python
# Preprocessing -> tokenisation, stop words, html cleaning

stop_words = set(stopwords.words('english'))

processed_review = []
# Go through each review and clean it
for review in data['review']:

  # remove html tags
  review = re.sub('<.*?>', ' ', review)
  # remove special characters
  review = re.sub('\W', ' ', review)
  # shorten spaces to one
  review = re.sub('\s+',' ', review)
  # remove stopwords
  word_tokens = word_tokenize(review)
  # join it together
  filtered_sentence = " ".join([w for w in word_tokens if w not in stop_words])

  # final process
  processed_review.append(filtered_sentence)

#print(processed_review[1])
```

```python
# Split the dataset into train/test
scaler = StandardScaler()
x_train, x_test, y_train, y_test = train_test_split(processed_review, data['sentiment'], test_size=0.25, random_state=42)

y_train = (y_train.replace({'positive': 1, 'negative': 0})).values
y_test = (y_test.replace({'positive': 1, 'negative': 0})).values
```

```python
# Neural Network model
# Naive Bayes (for comparison)

# Creating a TF-IDF vectorizer object
vectorizer = TfidfVectorizer()

# Fitting the vectorizer to the training data and transforming it
x_train_vec = vectorizer.fit_transform(x_train)

# Transforming the test data using the fitted vectorizer
x_test_vec = vectorizer.transform(x_test)


MNB = MultinomialNB()
MNB.fit(x_train_vec, y_train)
MNB_predict = MNB.predict(x_test_vec)
```

```python
# Neural Network model
# Sequential, LTSM

# 1. Tokenize the text data
tokenizer = Tokenizer(num_words=10000)  # Limit vocabulary size
tokenizer.fit_on_texts(x_train)
x_train_sequences = tokenizer.texts_to_sequences(x_train)
x_test_sequences = tokenizer.texts_to_sequences(x_test)

# 2. Pad sequences to ensure uniform length
max_length = 100  # Adjust as needed for timesteps
x_train_padded = pad_sequences(x_train_sequences, maxlen=max_length)
x_test_padded = pad_sequences(x_test_sequences, maxlen=max_length)

# 3. Create model
model = keras.Sequential([])
model.add(Embedding(10000, 128, input_length=max_length)) # Starting point

# Hidden Layers START
model.add(layers.LSTM(128, dropout=0.2, recurrent_dropout=0.2, return_sequences=True))
model.add(layers.LSTM(128, dropout=0.2, recurrent_dropout=0.2, return_sequences=True))
model.add(layers.LSTM(128, dropout=0.2, recurrent_dropout=0.2))
# Hidden Layers END

model.add(layers.Dense(1, activation='sigmoid')) # End point

# 4. Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 5. Train the model
#model.fit(x_train, y_train, batch_size=32, epochs=10, validation_data=(x_test, y_test))
#model.fit(x_train_padded, y_train, batch_size=32, epochs=10, validation_data=(x_test_padded, y_test))
model.fit(x_train_padded, y_train, batch_size=128, epochs=3, validation_data=(x_test_padded, y_test))
```

```python
# Evaluation (Classification Report)
# For Naive Baynes
nb_acc = accuracy_score(MNB_predict, y_test)

nb_class = classification_report(y_test, MNB_predict, target_names=['Negative','Positive'])
print(f"Accuracy Score:  {nb_acc}")

print(f"{nb_class}" )
```

```python
# Evaluation (Classification Report)
# For LTSM
ltsm_predict = model.predict(x_test_padded)
ltsm_predict = (ltsm_predict > 0.5).astype(int)

ltsm_acc = accuracy_score(y_test, ltsm_predict)

ltsm_class = classification_report(y_test, ltsm_predict)
print(f"Accuracy Score: {ltsm_acc}")

print(f"{ltsm_class}")
```

```python
# Visualisation (Confusion Matrix)
# for Naive Baynes
print("Confusion Matrix: \n", confusion_matrix(y_test, MNB_predict))
```

```python
# Visualisation (Confusion Matrix)
# for LTSM
print("Confusion Matrix: \n", confusion_matrix(y_test, ltsm_predict))
```
